You will use two modules that provide pseudo-random number generators to implement random sampling routines. Have a look [here](https://docs.python.org/3/library/random.html) (random module) and [here](https://numpy.org/doc/stable/reference/random/index.html) (numpy.random module). Both allow to simulate data and take random samples, although np.random allows more pseudo-random generator methods to be implemented.

To run the simulations, you should first import pandas, numpy and random modules (in case you still did not installed these modules, you will need to install beforehand).


```python
import pandas as pd
import numpy as np
import random
```

```python
# Simulate var1
var1 = []
random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
levels = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]
for _ in range(100): # a loop is needed because random.sample selects unique elements (with no replacement)
    var1 += random.sample(levels, 1) # var1.append(random.sample(levels, 1)) would also work
print(var1)
```

```python
# alternative to run a random sampling with replacement (using numpy)
levels = np.array(["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland"])
sampler = np.random.randint(0, len(levels), 100) # 100 random values within an interval (0 to 7)
var1 = levels.take(sampler) # use sampler to select values from "levels"; take - returns elements from array along the mentioned axis and indices
print(var1)
```

```python
sampler = np.random.randint(0, len(levels), 100)
type(sampler)
```

```python
# Simulate var2
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
var2 = np.random.uniform(0, 100, 100)
print(var2)
```

```python
# Simulate table1
table1 = pd.DataFrame(var1).value_counts(sort=True)
table1 = table1.rename_axis("landuse")
table1 = table1.reset_index(name="Frequency")
print(table1)
```

Table 1
Best type of table: Bar chart
Justification: The landuse is categorical, and so frequency displays clearly on a bar chart. 

```python
import matplotlib.pyplot as plt


plt.bar(table1["landuse"], table1["Frequency"])
plt.xlabel('Landuse')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Simulate table2
table2 = pd.DataFrame(list(zip(var1, var2)), columns = ["landuse", "cover"])
print(table2)
```

Table 2
Best type of table: A categorical scatterplot
Justification: A categorical scatterplot shows each data point for each landuse. 

```python
plt.scatter(table2["landuse"], table2["cover"])
plt.xticks(rotation=90)
plt.show()
```

```python
# Simulate table3
np.random.seed(24) # optional: used to fix the seed of the pseudo-random number generator (use any number of your choice)
year = list(range(1970,2021))
temp = np.random.normal(17,2,51)
table3 = pd.DataFrame(list(zip(year, temp)), columns = ["Year", "Temperature"])
print(table3)
```

Table 3
Best type of table: Linechart
Justification: A linechart works well for this data as it shows a trend over time. This data only has one data point per year, which is not continuous and so could make this confusing if not made clear, due to the nature of weather

```python
plt.plot(table3["Year"], table3["Temperature"])
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show()
```

```python
# Simulate table4
xx = np.array([16,21])
yy = np.array([300, 1200])
means = [xx.mean(), yy.mean()]  
stds = [xx.std() / 3, yy.std() / 3]
corr = -0.7 # correlation
covs = [[stds[0]**2          , stds[0]*stds[1]*corr], 
        [stds[0]*stds[1]*corr,           stds[1]**2]] # covariance matrix
table4 = pd.DataFrame(np.random.multivariate_normal(means, covs, 100), columns = ["Mean Anual Temperature", "Total Precipitation"])
print(table4)
```

Table 4
Best type of table: Scatterplot
Justification: A scatterplot makes it easy to see the connection between the two variables for each datapoint. It is also useful for seeing any potential correlation. For this graph it appears that their is a negative correlation between mean annual temperature and total precipitation i.e. as temperature goes up, rainfall goes down, which makes sense as the correlation is set to -0.7

```python
plt.scatter(table4["Mean Anual Temperature"], table4["Total Precipitation"])
plt.xlabel('Mean Anual Temperature')
plt.ylabel('Total Precipitation')
plt.show()
```

```python
# Simulate table5
col1 = pd.Series(list(range(1900,2010,10))).repeat(8)
col2 = ["Permanent crops", "Irrigated crops", "Managed Forest", "Natural Forest", "Agro-Forestry system", "Urban", "Pasture", "Shrubland" ]*11
col3 = np.random.uniform(0, 100, 90)
table5 = pd.DataFrame(list(zip(col1, col2, col3)), columns = ["Year", "Landuse", "Cover"])
print(table5)
```

Table 5
Best type of table: depends on what this data means. Either a categorical line chart or categorical scatterplot with different colours per category
Justification: It's unclear if the cover is for the same areas per year. If it is for the same area and is changing each year (e.g. landcover in Portugal) line chart would be better, with a line for each landuse, but as this is unclear and might just be the areas surveyed on that year, a scatterplot makes more sense as there is not necessarily a connection between each year's datapoints. As can be seen below, the scatterplot is a lot easier to look at than the line chart.

```python
# change landuse into numbers
table5['code'] = pd.factorize(table5['Landuse'], sort=True)[0] + 1 

plt.scatter(table5["Year"], table5["Cover"], c=table5["code"])
plt.xlabel('Year')
plt.ylabel('Cover')
plt.show()
```
